```python
#Bar chart for How Many passenger are monthly and Weekly Traveling. 
plt.figure(figsize=(16,8))
plt.bar('Ticket_Type','Ticket Fare', data=Train)
plt.ylabel('Ticket Fare')
plt.xlabel('Ticket_Type')
plt.title('How Many passenger are monthly and Weekly Traveling', fontsize = 20)
plt.show()
```

```python
#Total Number of Staion and Which Staion Most Passenger Travelling.
df_state_dealings = Train.groupby('Station Location')['Person_Arrived Location'].count().sort_values(ascending = False).plot.bar(figsize = (15, 5))
plt.ylabel('End_Station')
plt.xlabel('Starting_Station')
plt.title('Total Number of Station and Which Station Most Passenger Travelling ', fontsize = 20)
plt.show()
```

```python
# Creating a correlation matrix.
attributes = ['Starting_Station ','End_Station','Station Location','Person_Arrived Location']
corr_mat = Train.corr()
corr_mat
```

```python
plt.subplots(figsize=(12,9))
sns.heatmap(corr_mat,annot=True)
plt.title('Correlation, Starting_Station, End_Station, Station Location, Person_Arrived Location', fontsize = 20)
plt.show()
```
